```python
import pandas as pd
import numpy as np
```

```python
routes_df = pd.read_csv('routes.csv')
aircraft_df = pd.read_csv('aircraft_data_modified.csv')
planes_df = pd.read_csv('planes.dat.csv', header = None)

routes_df.head()
```

```python
aircraft_df.head()
```

```python
planes_df.head()
```

```python
merged_aircraft_planes = pd.merge(aircraft_df, planes_df, left_on='ICAO_code', right_on=3)
merged_aircraft_planes.drop(columns=[3,0], inplace=True)

merged_aircraft_planes.columns=['long_plane_code', 'classification', 'capacity', 'plane_type', 'short_code']
merged_aircraft_planes.head()
```

```python
list(routes_df[' equipment'].unique())
```

```python
routes_df.columns = ['airline', 'airline_ID', 'source_airport', 'source_airport_ID', 'destination_airport', 'destination_airport_ID','codeshare', 'stops', 'short_plane_code']
routes_df.head()
```

```python
routes_df['short_code'] = routes_df['short_plane_code'].str.split().str[-1].str[-3:]

routes_df.head()
```

```python
list(routes_df['short_code'].unique())
```

```python
final_df = pd.merge(routes_df, merged_aircraft_planes[['short_code', 'capacity']], how='left', on='short_code')

print(final_df)


```

```python
final_df.info()
```

```python
(final_df['capacity'].value_counts())
```

```python
capacity_inserts = {
    '146': 27,
    'CRJ': 70,
    '330': 260,
    'CNC': 12,
    'DH8': 40,
    'BET': 10,
    '73M': 149,
    '737': 189,
    'YK2': 104,
    'YK4': 25,
    '313': 200,
    '777': 375,
    'CNA': 7,
    'BE1': 10,
    'CNT': 10,
    'MA6': 65
}

capacities = []

final_df.dropna(subset='short_code', inplace=True)

final_df['capacity'].fillna(0, inplace=True)
count = 0

for row in final_df.iterrows():
    if row[1]['capacity'] != 0:
        capacities.append(row[1]['capacity'])
        count += 1
    else:
        try:
            capacities.append(capacity_inserts[row[1]['short_code']])
        except:
            capacities.append(np.median(final_df['capacity'].dropna()))
        count += 1

        
print(count)
        
```

```python
final_df['capacity'] = capacities
```

```python
final_df.info()
```

```python
final_df.to_csv('routes_w_capacities.csv', index=False)
```
